```python
# add algorithm column
sl_df['algorithm'] = 'Simple_LAP_Tracker'
lm_df['algorithm'] = 'Linear_motion_LAP_tracker'
lap_df['algorithm'] = 'LAP_Tracker'
nns_df['algorithm'] = 'Nearest_neighbor_search'
ol_df['algorithm'] = 'Old_LAP_Tracker'
osl_df['algorithm'] = 'Old_Simple_LAP_tracker'
```

```python
# create joined df with Simple_LAP_Tracker and Linear_motion_LAP_tracker median track speed data
sl_df['algorithm'] = 'Simple_LAP_Tracker'

sl_lm_df = pd.concat([sl_df, lm_df])
sl_lm_df = sl_lm_df.reset_index()
sl_lm_df
```

```python
sl_lm_TRACK_MEDIAN_SPEED_df = sl_lm_df[['TRACK_MEDIAN_SPEED', 'algorithm']]
sl_lm_TRACK_MEDIAN_SPEED_df.reset_index()

sns.displot(sl_lm_TRACK_MEDIAN_SPEED_df, x='TRACK_MEDIAN_SPEED', hue='algorithm', kind="kde", bw_adjust=.3, height=5, aspect=15/8)
```

```python
# concatenate all df

full_df = pd.concat([sl_df, lm_df, lap_df, nns_df, ol_df, osl_df])
full_df = full_df.reset_index()

full_TRACK_MEDIAN_SPEED_df = full_df[['TRACK_MEDIAN_SPEED', 'algorithm']]
```

```python
# TRACK_MEDIAN_SPEED histogram for all algorithms

sns.displot(full_TRACK_MEDIAN_SPEED_df, x='TRACK_MEDIAN_SPEED', height=5, aspect=15/8)
```

```python
# TRACK_MEDIAN_SPEED kernel density estimate for all algorithms

sns.displot(full_TRACK_MEDIAN_SPEED_df, x='TRACK_MEDIAN_SPEED', hue='algorithm', kind="kde", bw_adjust=.3, height=5, aspect=15/8)
```

```python
# TRACK_MEAN_SPEED kernel density estimate for all algorithms

full_TRACK_MEAN_SPEED_df = full_df[['TRACK_MEAN_SPEED', 'algorithm']]
sns.displot(full_TRACK_MEAN_SPEED_df, x='TRACK_MEAN_SPEED', hue='algorithm', kind="kde", bw_adjust=.3, height=5, aspect=15/8)
```

```python
# TRACK_STD_SPEED kernel density estimate for all algorithms

full_TRACK_STD_SPEED_df = full_df[['TRACK_STD_SPEED', 'algorithm']]
sns.displot(full_TRACK_STD_SPEED_df, x='TRACK_STD_SPEED', hue='algorithm', kind="kde", bw_adjust=.3, height=5, aspect=15/8)
```

>All three KDE distributions appear to show two seperate categories of algorithm:
<ol>
    <li>high density for track speeds between 0-6 pixels/frame
    <li>low density in that same range
</ol>

>This suggests that the four high density algorithms (Simple LAP, Linear Motion, Nearest Neighbor, Old simple LAP)
<br>
> capture cell speed differently from the two low density algorithms (LAP, Old LAP)

```python
# NUMBER_SPOTS kernel density estimate for all algorithms

full_NUMBER_SPOTS_df = full_df[['NUMBER_SPOTS', 'algorithm']]
sns.displot(full_NUMBER_SPOTS_df, x='NUMBER_SPOTS', hue='algorithm', kind="kde", bw_adjust=.3, height=5, aspect=15/8)
```

```python
# NUMBER_SPOTS vs. TRACK_MEAN_SPEED bivariate kernel density estimate for all algorithms

full_NUMBER_SPOTS_TRACK_MEAN_SPEED_df = full_df[['NUMBER_SPOTS', 'TRACK_MEAN_SPEED', 'algorithm']]
sns.displot(full_NUMBER_SPOTS_TRACK_MEAN_SPEED_df, x='TRACK_MEAN_SPEED', y='NUMBER_SPOTS', hue='algorithm', height=5, aspect=15/8)
```

```python
# NUMBER_SPOTS vs. TRACK_MEAN_SPEED jointplot for all algorithms

full_NUMBER_SPOTS_TRACK_MEAN_SPEED_df = full_df[['NUMBER_SPOTS', 'TRACK_MEAN_SPEED', 'algorithm']]
sns.jointplot(data=full_NUMBER_SPOTS_TRACK_MEAN_SPEED_df, x='TRACK_MEAN_SPEED', y='NUMBER_SPOTS', hue='algorithm', height=10)
```

```python
# NUMBER_SPOTS vs. TRACK_MEAN_SPEED Empirical cumulative distribution for all algorithms

full_TRACK_MEAN_SPEED_df = full_df[['TRACK_MEAN_SPEED', 'algorithm']]
sns.displot(full_TRACK_MEAN_SPEED_df, x='TRACK_MEAN_SPEED', hue='algorithm', kind='ecdf', height=8)
```
