```python
# compare mean velocity values found
# read in all csv files

# Simple_LAP_Tracker
sl_df = pd.read_csv('/Users/jonathanlifferth/Documents/Arbeit/Vanderbilt_Research_Analyst/Naotoshi_Nakamura/Baxter_experiments/JHH4_A2_trackmate_results/0_Simple LAP Tracker/Track statistics.csv')

# Linear_motion_LAP_tracker
lm_df = pd.read_csv('/Users/jonathanlifferth/Documents/Arbeit/Vanderbilt_Research_Analyst/Naotoshi_Nakamura/Baxter_experiments/JHH4_A2_trackmate_results/1_Linear motion LAP tracker/Track statistics.csv')

# LAP_Tracker
lap_df = pd.read_csv('/Users/jonathanlifferth/Documents/Arbeit/Vanderbilt_Research_Analyst/Naotoshi_Nakamura/Baxter_experiments/JHH4_A2_trackmate_results/2_LAP Tracker/Track statistics.csv')

# Nearest_neighbor_search
nns_df = pd.read_csv('/Users/jonathanlifferth/Documents/Arbeit/Vanderbilt_Research_Analyst/Naotoshi_Nakamura/Baxter_experiments/JHH4_A2_trackmate_results/3_Nearest neighbor search/Track statistics.csv')

# Old_LAP_Tracker
ol_df = pd.read_csv('/Users/jonathanlifferth/Documents/Arbeit/Vanderbilt_Research_Analyst/Naotoshi_Nakamura/Baxter_experiments/JHH4_A2_trackmate_results/4_Old LAP Tracker/Track statistics.csv')

# Old_Simple_LAP_tracker
osl_df = pd.read_csv('/Users/jonathanlifferth/Documents/Arbeit/Vanderbilt_Research_Analyst/Naotoshi_Nakamura/Baxter_experiments/JHH4_A2_trackmate_results/5_Old Simple LAP tracker/Track statistics.csv')
```

```python
# assign tracker and color lists
trackers = ['Simple LAP Tracker', 'Linear motion LAP tracker', 'LAP Tracker', 'Nearest neighbor search', 'Old LAP Tracker', 'Old Simple LAP tracker']
colors = ['b', 'g', 'r', 'c', 'y', 'k']
```

```python
# calculate total mean of mean track speed for each alogorithm
sl_df_mean = np.mean(sl_df.TRACK_MEAN_SPEED)
lm_df_mean = np.mean(lm_df.TRACK_MEAN_SPEED)
lap_df_mean = np.mean(lap_df.TRACK_MEAN_SPEED)
nns_df_mean = np.mean(nns_df.TRACK_MEAN_SPEED)
ol_df_mean = np.mean(ol_df.TRACK_MEAN_SPEED)
osl_df_mean = np.mean(osl_df.TRACK_MEAN_SPEED)
# baxter_mean_speed = 

mean_speeds = [sl_df_mean, lm_df_mean, lap_df_mean, nns_df_mean, ol_df_mean, osl_df_mean]
```

```python
sl_df_track_count = sl_df.shape[0]
lm_df_track_count = lm_df.shape[0]
lap_df_track_count = lap_df.shape[0]
nns_df_track_count = nns_df.shape[0]
ol_df_track_count = ol_df.shape[0]
osl_df_track_count = osl_df.shape[0]
baxter_track_count = 110

track_counts = [sl_df_track_count, lm_df_track_count, lap_df_track_count, nns_df_track_count, ol_df_track_count, osl_df_track_count]
```

```python
plt.figure(figsize=(10,5))

for i in range(len(trackers)):
    plt.scatter(track_counts[i], mean_speeds[i], c=colors[i], label=trackers[i],
                s=250, alpha=0.7)

plt.title('Performance of tracking algorithms for JHH4_A2 cell line')
plt.xlabel('Track count')
plt.ylabel('Mean of mean track speeds')
plt.legend()
plt.grid(True)
# figure(figsize=(10, 10), dpi=100)
# plt.rcParams["figure.figsize"] = (5,5)
plt.show()
```
