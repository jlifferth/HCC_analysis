# Individual Cell Line Distributions
Comparison of mean track speed distribution between each image series for each cell line

1. Huh7
2. JHH4
3. JHH6
4. JHH7
5. Kami41
6. PRF5

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
sns.set_context("talk")

master_df = pd.read_csv('master_df_2.csv')

master_df.IMAGE_SERIES.unique()
```

```python
Huh7_lines = ['Huh7_B4_Pickup', 'Huh7_A4', 'Huh7_C4', 'Huh7_D4']

Huh7_df = master_df[master_df.IMAGE_SERIES.isin(Huh7_lines)]
Huh7_df = Huh7_df.reset_index()

Huh7_TRACK_MEAN_SPEED_df = Huh7_df[['TRACK_MEAN_SPEED', 'IMAGE_SERIES']]

sns.displot(Huh7_TRACK_MEAN_SPEED_df, x='TRACK_MEAN_SPEED', hue='IMAGE_SERIES', kind="kde", bw_adjust=.3, height=5, aspect=15/8)
```

```python
JHH4_lines = ['JHH4_A2', 'JHH4_B2', 'JHH4 C2 Pickup', 'JHH4 D2']

JHH4_df = master_df[master_df.IMAGE_SERIES.isin(JHH4_lines)]
JHH4_df = JHH4_df.reset_index()

JHH4_TRACK_MEAN_SPEED_df = JHH4_df[['TRACK_MEAN_SPEED', 'IMAGE_SERIES']]

sns.displot(JHH4_TRACK_MEAN_SPEED_df, x='TRACK_MEAN_SPEED', hue='IMAGE_SERIES', kind="kde", bw_adjust=.3, height=5, aspect=15/8)
```

```python
JHH6_lines = ['JHH6_A1', 'JHH6_B1', 'JHH6_C1_pickup', 'JHH6_D1']

JHH6_df = master_df[master_df.IMAGE_SERIES.isin(JHH6_lines)]
JHH6_df = JHH6_df.reset_index()

JHH6_TRACK_MEAN_SPEED_df = JHH6_df[['TRACK_MEAN_SPEED', 'IMAGE_SERIES']]

sns.displot(JHH6_TRACK_MEAN_SPEED_df, x='TRACK_MEAN_SPEED', hue='IMAGE_SERIES', kind="kde", bw_adjust=.3, height=5, aspect=15/8)
```

```python
JHH7_lines = ['JHH7_C5', 'JHH7_D5', 'JHH7_A5', 'JHH7_B5_pickup']

JHH7_df = master_df[master_df.IMAGE_SERIES.isin(JHH7_lines)]
JHH7_df = JHH7_df.reset_index()

JHH7_TRACK_MEAN_SPEED_df = JHH7_df[['TRACK_MEAN_SPEED', 'IMAGE_SERIES']]

sns.displot(JHH7_TRACK_MEAN_SPEED_df, x='TRACK_MEAN_SPEED', hue='IMAGE_SERIES', kind="kde", bw_adjust=.3, height=5, aspect=15/8)
```

```python
Kami41_lines = ['Kami41_A6', 'Kami41_C6', 'Kami41_D6', 'Kami41_B6_pickup']

Kami41_df = master_df[master_df.IMAGE_SERIES.isin(Kami41_lines)]
Kami41_df = Kami41_df.reset_index()

Kami41_TRACK_MEAN_SPEED_df = Kami41_df[['TRACK_MEAN_SPEED', 'IMAGE_SERIES']]

sns.displot(Kami41_TRACK_MEAN_SPEED_df, x='TRACK_MEAN_SPEED', hue='IMAGE_SERIES', kind="kde", bw_adjust=.3, height=5, aspect=15/8)
```

```python
PRF5_lines = ['PRF5_B3_pickup', 'PRF5_A3', 'PRF5_C3', 'PRF5_D3']

PRF5_df = master_df[master_df.IMAGE_SERIES.isin(PRF5_lines)]
PRF5_df = PRF5_df.reset_index()

PRF5_TRACK_MEAN_SPEED_df = PRF5_df[['TRACK_MEAN_SPEED', 'IMAGE_SERIES']]

sns.displot(PRF5_TRACK_MEAN_SPEED_df, x='TRACK_MEAN_SPEED', hue='IMAGE_SERIES', kind="kde", bw_adjust=.3, height=5, aspect=15/8)
```

```python
# add cell line column
Huh7_df['cell_line'] = 'Huh7'
JHH4_df['cell_line'] = 'JHH4'
JHH6_df['cell_line'] = 'JHH6'
JHH7_df['cell_line'] = 'JHH7'
Kami41_df['cell_line'] = 'Kami41'
PRF5_df['cell_line'] = 'PRF5'
```

```python
# concatenate all df

full_df = pd.concat([Huh7_df, JHH4_df, JHH6_df, JHH7_df, Kami41_df, PRF5_df])
full_df = full_df.reset_index(drop=True)

full_df.to_csv('full_df.csv')
```
