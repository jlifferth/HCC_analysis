# Example 1: One Sample Kolmogorov-Smirnov Test 

```python

#set seed (e.g. make this example reproducible)
seed(0)

#generate dataset of 100 values that follow a Poisson distribution with mean=5
data = poisson(5, 100)
```

```python
sns.displot(data, kind='ecdf')
sns.displot(data, kind='kde', bw_adjust=.4)
sns.displot(data)
sns.displot(data, kde=True)
```

```python
from scipy.stats import kstest
```

```python
#perform Kolmogorov-Smirnov test
kstest(data, 'norm')

# KstestResult(statistic=0.9072498680518208, pvalue=1.0908062873170218e-103)
```

```python
def calc_ks(data):
    series_list = data.columns
    
    ks_values = []
    
    # calculate KS values
    for i in range(len(series_list)):
        results_list = [series_list[i]]
        for j in range(len(series_list)):
            p = series_list[i]
            q = series_list[j]
            comparison = p + " x " + q
            p = data[p]
            q = data[q]
            ks_result = ks_2samp(p, q)[0]
            results_list.append(ks_result)

        ks_values.append(results_list)
    
    print(ks_values)
    
    # convert to dataframe
    df = pd.DataFrame(ks_values, columns = ['Index', series_list[0], series_list[1], series_list[2], 
                                            series_list[3]])
    df = df.set_index('Index')
    print(df)
    
    with sns.axes_style("white"):
        mask = np.zeros_like(df)
        mask[np.triu_indices_from(mask)] = True    
        sns.heatmap(df, annot=True, mask=mask, vmin=0, vmax=0.5, cmap="YlGnBu")
    
calc_ks(PRF5_bootstrap_df)
```

```python
def distribution_and_ks(data, cell_line):
    bootstrap_dict = perform_bootstrap(dataframe=data, cell_line=cell_line)
    bootstrap_df = pd.DataFrame.from_dict(bootstrap_dict)
    bootstrap_df = bootstrap_df.reset_index(drop=True)
    bootstrap_df
    
    # calculate Kolmogorov-Smirnov
    series_list = bootstrap_df.columns
    ks_values = []
    p_values = []
    # calculate KS values
    for i in range(len(series_list)):
        results_list = [series_list[i]]
        for j in range(len(series_list)):
            p = series_list[i]
            q = series_list[j]
            comparison = p + " x " + q
            p = bootstrap_df[p]
            q = bootstrap_df[q]
            print('Full KS sample result: ', ks_2samp(p, q))
            ks_result = ks_2samp(p, q)[0]
            p_value = ks_2samp(p, q)[1]
            print('p_value: ', p_value)
            results_list.append(ks_result)

        ks_values.append(results_list)
    
    # convert to dataframe
    ks_df = pd.DataFrame(ks_values, columns = ['Index', series_list[0], series_list[1], series_list[2], 
                                            series_list[3]])
    ks_df = ks_df.set_index('Index')
    # plot distribution
    sns.displot(bootstrap_df, kind="kde", bw_adjust=.3, 
                height=5, aspect=15/8).set(title='{} distribution with bootstrap'.format(cell_line))
#     plt.savefig('./images/{}_kde.png'.format(cell_line), dpi=200, bbox_inches='tight', pad_inches=0.1)
    plt.show()
    
    with sns.axes_style("white"):
        mask = np.zeros_like(ks_df)
        mask[np.triu_indices_from(mask)] = True    
        sns.heatmap(ks_df, annot=True, mask=mask, vmin=0, vmax=0.5, 
                    cmap="YlGnBu").set(title='{} Kolmogorov-Smirnov matrix'.format(cell_line))
#         plt.savefig('./images/{}_ks_heatmap.png'.format(cell_line), dpi=200, bbox_inches='tight', pad_inches=0.1)
    plt.show()


distribution_and_ks(data=df, cell_line='JHH4')
```

```python
all_ks_and_p = [] # this will be a list of series with min_max values for each cell line

def calc_ks_and_p(data, cell_line):
    bootstrap_dict = perform_bootstrap(dataframe=data, cell_line=cell_line)
    bootstrap_df = pd.DataFrame.from_dict(bootstrap_dict)
    bootstrap_df = bootstrap_df.reset_index(drop=True)
    bootstrap_df
    
    # calculate Kolmogorov-Smirnov
    series_list = bootstrap_df.columns
    ks_values = []
    p_values = []
    # calculate KS values
    for i in range(len(series_list)):
        ks_list = [series_list[i]]
        p_value_list = [series_list[i]]
        for j in range(len(series_list)):
            p = series_list[i]
            q = series_list[j]
            comparison = p + " x " + q
            p = bootstrap_df[p]
            q = bootstrap_df[q]
            ks_result = ks_2samp(p, q)[0]
            p_value = ks_2samp(p, q)[1]
            ks_list.append(ks_result)
            p_value_list.append(p_value)

        ks_values.append(ks_list)
        p_values.append(p_value_list)
    
#     print(ks_values)
#     print(p_values)

    # convert to dataframe
    ks_df = pd.DataFrame(ks_values, columns = ['Index', series_list[0], series_list[1], series_list[2], 
                                            series_list[3]])
    ks_df = ks_df.set_index('Index')
    
    p_df = pd.DataFrame(p_values, columns = ['Index', series_list[0], series_list[1], series_list[2], 
                                            series_list[3]])
    p_df = p_df.set_index('Index')

    # remove =0 and p=1 (indicates pairwise KS on same sample)
    ks_df = ks_df.replace(0, np.nan)
    p_df = p_df.replace(1, np.nan)

    # find max and min ks  and p values
    
    ks_min = ks_df.min().min()
    ks_max = ks_df.max().max()
    p_min = p_df.min().min()
    p_max = p_df.max().max()
    
    min_max_series = pd.Series([ks_min, ks_max, p_min, p_max])
    min_max_series.name = cell_line
    
    all_ks_and_p.append(min_max_series)
    
    data = np.asarray(min_max_series).reshape(1,len(min_max_series))

    print(min_max_series)
    print(data)
    
#     print(ks_df.min().min())
#     print(ks_df.max().max())
#     print(p_df.min().min())
#     print(p_df.max().max())
    
    # plot distribution
#     sns.displot(bootstrap_df, kind="kde", bw_adjust=.3, 
#                 height=5, aspect=15/8).set(title='{} distribution with bootstrap'.format(cell_line))
#     plt.savefig('./images/{}_kde.png'.format(cell_line), dpi=200, bbox_inches='tight', pad_inches=0.1)
#     plt.show()
    
    with sns.axes_style("white"):
        mask = np.zeros_like(ks_df)
        mask[np.triu_indices_from(mask)] = True    
        sns.heatmap(ks_df, annot=True, mask=mask, vmin=0, vmax=0.5, 
                    cmap="YlGnBu").set(title='{} Kolmogorov-Smirnov -value matrix'.format(cell_line))
#         plt.savefig('./images/{}_ks_heatmap.png'.format(cell_line), dpi=200, bbox_inches='tight', pad_inches=0.1)
    plt.show()
    
    with sns.axes_style("white"):
        mask = np.zeros_like(p_df)
        mask[np.triu_indices_from(mask)] = True    
        sns.heatmap(p_df, annot=True, mask=mask, vmin=0, vmax=0.5, 
                    cmap="YlGnBu").set(title='{} Kolmogorov-Smirnov p-value matrix'.format(cell_line))
#         plt.savefig('./images/{}_ks_heatmap.png'.format(cell_line), dpi=200, bbox_inches='tight', pad_inches=0.1)
    plt.show()
    
    sns.heatmap(data, annot=True, vmin=0, vmax=0.5, 
                    cmap="YlGnBu").set(title='{} Kolmogorov-Smirnov min-max -/ p-values'.format(cell_line))


# calc_ks_and_p(data=df, cell_line='JHH4')
cell_lines = df.cell_line.unique()


for cell_line in cell_lines:
    calc_ks_and_p(data=df, cell_line=cell_line)
```

```python
series_names = [series.name for series in all_ks_and_p]

print(series_names)
print(len(all_ks_and_p))

all_ks_and_p_df = pd.DataFrame(all_ks_and_p)
all_ks_and_p_df.rename(columns={0: ' min', 1: ' max', 2:'p min', 3:'p max'}, inplace=True)
to_drop = ['JHH6_0uM', 'JHH6_1uM', 'JHH6_2.5uM', 'JHH7_0uM', 'JHH7_1uM', 'JHH7_2.5uM']
all_ks_and_p_df = all_ks_and_p_df.drop(to_drop)

all_ks_and_p_df
```
