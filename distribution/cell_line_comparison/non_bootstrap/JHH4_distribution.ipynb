```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
sns.set_context("paper")
```

```python
master_df = pd.read_csv('master_df_2.csv')

master_df.IMAGE_SERIES.unique()
```

```python
JHH4_lines = ['JHH4_A2', 'JHH4_B2', 'JHH4 C2 Pickup', 'JHH4 D2']

JHH4_df = master_df[master_df.IMAGE_SERIES.isin(JHH4_lines)]
JHH4_df = JHH4_df.reset_index()

JHH4_df
```

```python
JHH4_A2 = JHH4_df[JHH4_df['IMAGE_SERIES'] == 'JHH4_A2']
JHH4_A2 = JHH4_A2.reset_index()
JHH4_A2

JHH4_A2_TRACK_MEAN_SPEED_df = JHH4_A2['TRACK_MEAN_SPEED']
JHH4_A2_TRACK_MEAN_SPEED_df

sns.displot(JHH4_A2_TRACK_MEAN_SPEED_df)
```

```python
JHH4_TRACK_MEAN_SPEED_df = JHH4_df[['TRACK_MEAN_SPEED', 'IMAGE_SERIES']]
```

```python
sns.displot(JHH4_TRACK_MEAN_SPEED_df, x='TRACK_MEAN_SPEED', hue='IMAGE_SERIES', kind="kde", bw_adjust=.3, height=5, aspect=15/8)

```
